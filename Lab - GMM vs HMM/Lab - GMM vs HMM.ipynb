# Hidden Markov Model VS Gaussian Mixture Models

#### Fitting both models on data

```python
from PGM_HW3_tools import my_HMM
from PGM_HW3_tools import my_GMM
```

```python
HMM=my_HMM(n_clusters=4, initialization="KMeans", RandomState=1)
HMM.fit(X_data)

GMM=my_GMM(n_clusters=4)
GMM.fit(X_data)
```

#### Comparing results

```python
from PGM_HW3_tools import plot_results
```

```python
plot_results(X_data,X_test,HMM,HMM.labels_,HMM.predict_probable_sequence(X_test),'HMM')
plot_results(X_data,X_test,GMM,GMM.labels_,GMM.predict(X_test),'GMM')
```
